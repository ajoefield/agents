```python
from dotenv import load_dotenv


load_dotenv(override=True)
```

```python
# Check for bedrock api key
import os
import boto3

bedrock_api_key = os.getenv('AWS_BEARER_TOKEN_BEDROCK')

if bedrock_api_key:
    print(f"Bedrock API Key exists and begins {bedrock_api_key[:8]}")
else:
    print("Bedrock API Key not set - please head to the troubleshooting guide in the setup folder")
```

```python
# Create an Amazon Bedrock client
client = boto3.client(
    service_name="bedrock-runtime",
    region_name="us-west-2" # If you've configured a default region, you can omit this line
)
```

```python
# Define the model and message
model_id = "us.anthropic.claude-3-5-haiku-20241022-v1:0"
messages = [{"role": "user", "content": [{"text": "Please propose a hard, challenging question to assess someone's IQ. Respond only with the question."}]}]

print(messages)
```

```python
# Get repsonse with Bedrock

response = client.converse(
    modelId=model_id,
    messages=messages,
)

print(response)
user_friendly_output = response.get("output", {}).get("message", {}).get("content", [])[0].get("text")

print(user_friendly_output)
```
